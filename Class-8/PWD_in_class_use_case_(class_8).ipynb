```python
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns

print("Matplotlib and Seaborn are imported")
```

```python
sns.lmplot(x='area', y='price', data=df, height=10, scatter_kws={"s": 100}, hue='furnishingstatus')
```

```python
X = df[['bedrooms','bathrooms','stories','area']]
y = df[['price']]
```

```python
# Provide your solution here

from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

linreg.fit(X, y)
```

```python
linreg.predict([[3,1,2,13200]])
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df.mainroad)
df['mainroad'] = le.transform(df.mainroad)
```

```python
df.head(10)
```

```python
X = df[['bedrooms','bathrooms','stories','area','mainroad']]
y = df[['price']]

from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

linreg.fit(X, y)

linreg.predict([[3,1,2,13200,1]])
```
